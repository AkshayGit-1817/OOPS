1. What is Object-Oriented Programming (OOP)?
:-  
   Object-Oriented Programming is a software design model that structures code around data (objects) rather than functions and step-by-step logic. It groups related data and behaviors into self-contained building blocks to make complex programs modular, reusable, and easy to maintain.
2. What is a class in OOP?
:-  
  A class is an abstract blueprint, template, or prototype used to create objects. It defines the variables (attributes) and functions (methods) that all objects created from it will possess.
3. What is an object in OOP?
:-
   An object is a concrete, real-world instance of a class. If Car is the structural blueprint (the class), a specific red sedan parked in your driveway is the object.

4. What is the difference between abstraction and encapsulation?
:-
  * Abstraction is about hiding complexity by only showing essential features to the user (e.g., you press an accelerator pedal without needing to see how the engine combustion works).
  
  * Encapsulation is about hiding data by bundling data and methods inside a single unit and restricting direct access to protect it from outside tampering.
  
5. What are dunder methods in Python?
:-
  Dunder (Double Underscore) methods—also called magic methods—are special built-in methods surrounded by double underscores (e.g., __init__, __str__). They allow you to define how your custom objects behave with standard Python operators and built-in functions (+, ==, len(), print).

6. Explain the concept of inheritance in OOP.
:-
 Inheritance allows a new class (the child/derived class) to automatically adopt the attributes and methods of an existing class (the parent/base class). It promotes code reusability and establishes a natural "is-a" hierarchy (e.g., a Dog is-a Animal).

7. What is polymorphism in OOP?
:-
  Polymorphism means "many forms." It allows different objects to respond to the exact same method call in their own unique ways. For example, calling .make_sound() yields a bark for a Dog object but a meow for a Cat object.

8. How is encapsulation achieved in Python?
:-
  Python achieves encapsulation via naming conventions to signify access levels, as it lacks strict private keywords:
  
  * Protected: Single underscore prefix (_variable). It warns developers the data is intended for internal use.
  
  * Private: Double underscore prefix (__variable). It triggers name mangling to actively prevent direct accidental outside access.

9. What is a constructor in Python?
:-
   A constructor is a special method automatically called when a new object is instantiated. In Python, the __init__ method serves as the constructor, primarily used to assign initial values to an object's attributes.

10. What are class and static methods in Python?
:-
 * Class Methods (@classmethod): Methods that take the class itself (cls) as the first argument. They can access and modify class-level state.

 * Static Methods (@staticmethod): Methods bound to a class that do not take a mandatory first argument (self or cls). They act like normal functions isolated inside the class namespace.

11. What is method overloading in Python?
:-
Method overloading is the ability to define multiple methods with the same name but different input parameters. Python does not natively support traditional method overloading; defining a duplicate method name simply overwrites the previous one. Instead, Python handles this using default arguments or variable positional arguments (*args).

12. What is method overriding in OOP?
:-
  Method overriding occurs when a child class provides a specific implementation of a method that is already defined in its parent class. This allows the child class to alter or completely replace the inherited behavior.

13. What is a property decorator in Python?
:-
  The @property decorator allows you to turn a class method into a read-only attribute getter. It lets you access a method's return value like a regular variable (e.g., user.fullname instead of user.fullname()), enabling clean data validation behind the scenes.

14. Why is polymorphism important in OOP?
:-
  Polymorphism allows you to write highly flexible, generic code. You can create a single function that processes a list of mixed objects and triggers their respective behaviors accurately, without needing complex conditional loops (if/else) to check every object's specific type.

15. What is an abstract class in Python?
:-
  An abstract class is a conceptual class that cannot be instantiated directly. Created using Python’s abc module, it contains abstract methods that act as a mandatory "to-do list" which any child class must implement before it can be used.

16. What are the advantages of OOP?
:-
 * Code Reusability: Inheriting existing code avoids redundancy.
 * Easier Troubleshooting: Modular objects make bugs simple to isolate.
 * Scalability: Large applications can be expanded predictably by introducing new classes.
 * Data Security: Encapsulation protects sensitive application data.

17. What is the difference between a class variable and an instance variable?
:-
  * Class Variables: Defined directly inside the class body. They are shared equally by all instances of that class.
  * Instance Variables: Defined inside methods (usually __init__) using self.. They are completely unique to each individual object.

18. What is multiple inheritance in Python?
:-
  Multiple inheritance is a feature where a single child class can inherit directly from more than one parent class (e.g., class Smartphone inherits from both Phone and Camera). Python manages the execution order of these parents using the Method Resolution Order (MRO).

19. Explain the purpose of __str__ and __repr__ methods in Python ?
:-
  * __str__: Returns a user-friendly, readable string representation of an object, optimized for end-users (triggered by print() or str()).

  * __repr__: Returns an unambiguous, detailed string representation aimed at developers for debugging (triggered by running the object directly in an interactive console). It ideally mimics the code required to recreate the object.

20. What is the significance of the super() function in Python?
:-
  The super() function returns a temporary object of the parent class, allowing a child class to call methods and constructors from its parent. This prevents you from hardcoding parent class names and ensures clean cooperation during multiple inheritance.

21. What is the significance of the __del__ method in Python?
:-
  Known as the destructor, __del__ is invoked automatically when an object is about to be destroyed and garbage collected. It is typically used to perform cleanup actions, such as closing open network sockets, database connections, or active file pointers.

22. What is the difference between @staticmethod and @classmethod in Python?
:-
   * @classmethod accepts cls as its first parameter, allowing it to inspect or change class-wide attributes.
  
   * @staticmethod receives no automatic implicit arguments; it behaves exactly like an isolated utility function that merely lives inside the class's organizational boundary.

23. How does polymorphism work in Python with inheritance?
:-
  A child class inherits methods from its parent, but can override them to alter their actions. When you iterate through a collection of varying child objects and call the inherited method name, Python evaluates the underlying object type dynamically at runtime and runs the overridden method matching that specific child.

24. What is method chaining in Python OOP?
:-
Method chaining is a design pattern where multiple methods are called sequentially on a single object in a single line of code (e.g., car.accelerate().steer().brake()). This is accomplished by having each method end with a return self statement, passing the object's instance directly to the next call.

```python
1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog
that overrides the speak() method to print "Bark!".
:-
  class Animal:
    def speak(self):
        print("This animal makes a generic sound.")

class Dog(Animal):
    def speak(self):
        print("Bark!")

# --- Verification ---

# Create an instance of the parent class
generic_animal = Animal()
generic_animal.speak()  # Output: This animal makes a generic sound.

# Create an instance of the child class
my_dog = Dog()
my_dog.speak()          # Output: Bark!


2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.
:-
   from abc import ABC, abstractmethod
import math

# 1. Define the Abstract Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        """Calculates and returns the area of the shape."""
        pass

# 2. Derive the Circle Class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# 3. Derive the Rectangle Class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# --- Verification ---

# Trying to instantiate Shape directly would raise a TypeError:
# my_shape = Shape() # ❌ This will crash the program!

# Creating concrete instances instead:
shapes = [
    Circle(radius=5),
    Rectangle(width=4, height=6)
]

for shape in shapes:
    # Polymorphism in action: calling .area() behaves differently for each shape
    print(f"The area of the {type(shape).__name__} is: {shape.area():.2f}")


3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.
:-
   class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, doors):
        # Pass vehicle_type up to the Vehicle parent class
        super().__init__(vehicle_type)
        self.doors = doors

class ElectricCar(Car):
    def __init__(self, vehicle_type, doors, battery_capacity):
        # Pass vehicle_type and doors up to the Car parent class
        super().__init__(vehicle_type, doors)
        self.battery = battery_capacity

# --- Verification ---

# Instantiate the leaf (child) class of the multi-level chain
my_tesla = ElectricCar(vehicle_type="Sedan", doors=4, battery_capacity="85 kWh")

# Verify that the object possesses all attributes from the inheritance hierarchy
print(f"Vehicle Type:     {my_tesla.type}")     # Inherited from Vehicle
print(f"Number of Doors:  {my_tesla.doors}")    # Inherited from Car
print(f"Battery Capacity: {my_tesla.battery}")  # Defined in ElectricCar


4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.
:-
  class Bird:
    def fly(self):
        print("Most birds can fly high in the sky.")

class Sparrow(Bird):
    def fly(self):
        print("The sparrow chirps and flies short distances quickly.")

class Penguin(Bird):
    def fly(self):
        print("The penguin cannot fly! It uses its wings to swim instead.")

# --- Verification ---

# Create a heterogeneous list containing different types of birds
flock = [Sparrow(), Penguin(), Bird()]

# Polymorphism in action: Call the same method name on every object
for bird in flock:
    print(f"{type(bird).__name__}: ", end="")
    bird.fly()


5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.
:-
   class BankAccount:
    def __init__(self, owner, initial_balance=0.0):
        self.owner = owner
        # Private attribute: cannot be accessed directly outside the class
        self.__balance = float(initial_balance)

    def deposit(self, amount):
        """Safely increases the balance if the amount is valid."""
        if amount > 0:
            self.__balance += amount
            print(f"Successfully deposited ₹{amount:,.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        """Safely decreases the balance after checking funds availability."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self.__balance:
            print("Insufficient funds! Transaction denied.")
        else:
            self.__balance -= amount
            print(f"Successfully withdrew ₹{amount:,.2f}")

    def get_balance(self):
        """Provides controlled, read-only access to the private balance."""
        return self.__balance

# --- Verification ---

# Create an account
account = BankAccount(owner="Amit", initial_balance=1000.0)

# 1. Use the controlled public methods
account.deposit(500.0)      # Output: Successfully deposited ₹500.00
account.withdraw(200.0)     # Output: Successfully withdrew ₹200.00
print(f"Current Balance: ₹{account.get_balance():,.2f}")  # Output: ₹1,300.00

# 2. Try to break encapsulation
try:
    print(account.__balance)  # ❌ Attempt direct access
except AttributeError as e:
    print(f"\nEncapsulation active! Blocked direct access: {e}")

# 3. Try a bad transaction
account.withdraw(5000.0)    # Output: Insufficient funds! Transaction denied.


6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
and Piano that implement their own version of play().
:-
  class Instrument:
    def play(self):
        print("This instrument plays a generic musical note.")

class Guitar(Instrument):
    def play(self):
        print("The guitar strums an acoustic chord: *Pluck, Strum!*")

class Piano(Instrument):
    def play(self):
        print("The piano strikes a grand harmony: *Plink, Plonk!*")

# --- Verification ---

def perform(instrument_instance):
    """
    A unified function that accepts any Instrument family member.
    The exact code execution is decided at runtime.
    """
    instrument_instance.play()

# Create individual objects
generic_horn = Instrument()
my_fender = Guitar()
my_yamaha = Piano()

# Call the unified function with different objects
print("--- Band Performance ---")
perform(generic_horn)  # Output: This instrument plays a generic musical note.
perform(my_fender)     # Output: The guitar strums an acoustic chord: *Pluck, Strum!*
perform(my_yamaha)     # Output: The piano strikes a grand harmony: *Plink, Plonk!*


7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.
:-
  class MathOperations:
    # A class variable to keep track of operations performed (optional, for demonstration)
    operation_count = 0

    @classmethod
    def add_numbers(cls, a, b):
        """
        Class method: Receives the class ('cls') as the first implicit argument.
        Can access and modify class-level state.
        """
        cls.operation_count += 1
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        """
        Static method: Receives no implicit first argument (neither self nor cls).
        Behaves exactly like a standard, isolated function.
        """
        # Cannot access 'cls.operation_count' directly without explicitly naming the class
        return a - b

# --- Verification ---

# 1. Call methods directly using the Class Name (No instantiation required)
sum_result = MathOperations.add_numbers(15, 10)
diff_result = MathOperations.subtract_numbers(15, 10)

print(f"Addition Result: {sum_result}")       # Output: 25
print(f"Subtraction Result: {diff_result}")  # Output: 5

# 2. Verify class method side-effect
print(f"Total operations tracked by class: {MathOperations.operation_count}")  # Output: 1


8. Implement a class Person with a class method to count the total number of persons created.
:-
   class Person:
    # Class variable shared across all instances to track total population
    __total_count = 0

    def __init__(self, name):
        self.name = name
        # Increment the shared counter whenever a new instance is created
        Person.__total_count += 1

    @classmethod
    def get_total_count(cls):
        """
        Class method to safely access the private class variable.
        Receives 'cls' (the class blueprint itself) as its first argument.
        """
        return cls.__total_count

# --- Verification ---

# 1. Initial count check
print(f"Initial Count: {Person.get_total_count()}")  # Output: 0

# 2. Instantiate Person objects
p1 = Person("Alice")
p2 = Person("Bob")
p3 = Person("Charlie")

# 3. Check updated count via the class
print(f"Total Persons Created: {Person.get_total_count()}")  # Output: 3

# 4. You can also call it via an instance, but it still references class state
print(f"Count accessed via Alice: {p1.get_total_count()}")  # Output: 3


9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator".
:-
  class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        """
        Overrides the built-in string conversion behavior.
        Triggered automatically by print() or str().
        """
        return f"{self.numerator}/{self.denominator}"

# --- Verification ---

# 1. Create instances of the Fraction class
half = Fraction(1, 2)
three_quarters = Fraction(3, 4)

# 2. Print them directly to see the customized string output
print(f"First Fraction:  {half}")            # Output: 1/2
print(f"Second Fraction: {three_quarters}")  # Output: 3/4

# 3. Explicitly convert to a string variable
fraction_string = str(half)
print(f"String variable: {fraction_string}")   # Output: 1/2


10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.
:-
  class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        """
        Overloads the '+' operator.
        'self' represents the vector on the left side of the '+'.
        'other' represents the vector on the right side of the '+'.
        """
        # Return a brand new Vector containing the combined coordinates
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        # Human-readable format for verification
        return f"Vector({self.x}, {self.y})"

# --- Verification ---

# 1. Initialize two separate vector points
v1 = Vector(2, 4)
v2 = Vector(5, -1)

# 2. Use the standard '+' operator directly on the objects
v3 = v1 + v2

# 3. Print the result
print(f"Vector 1: {v1}")  # Output: Vector(2, 4)
print(f"Vector 2: {v2}")  # Output: Vector(5, -1)
print(f"Result (v1 + v2): {v3}")  # Output: Vector(7, 3)


11. Create a class Person with attributes name and age. Add a method greet() that prints "Hello, my name is
{name} and I am {age} years old."
:-
   class Person:
    def __init__(self, name, age):
        # Setting instance attributes unique to each created object
        self.name = name
        self.age = age

    def greet(self):
        """Prints a personalized greeting using the object's attributes."""
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# --- Verification ---

# 1. Create different individual instances of the Person class
person1 = Person("Aarav", 25)
person2 = Person("Ananya", 31)

# 2. Call the greet method on each object
person1.greet()  # Output: Hello, my name is Aarav and I am 25 years old.
person2.greet()  # Output: Hello, my name is Ananya and I am 31 years old.


12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades.
:-
   class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = list(grades)  # Ensures the input is stored as a list

    def average_grade(self):
        """Calculates and returns the arithmetic mean of the grades list."""
        if not self.grades:
            return 0.0  # Guard clause to handle DivisionByZero if no grades exist
        return sum(self.grades) / len(self.grades)

# --- Verification ---

# 1. Create a student instance with a list of marks
student1 = Student("Rohan", [85, 90, 78, 92])

# 2. Compute and print the average
avg = student1.average_grade()
print(f"Student Name: {student1.name}")
print(f"Grades:       {student1.grades}")
print(f"Average Mark: {avg:.2f}")  # Output: 86.25

# 3. Verification of empty edge-case handling
student2 = Student("Anjali", [])
print(f"\n{student2.name}'s average: {student2.average_grade():.1f}")  # Output: 0.0


13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the
area.
:-
   class Rectangle:
    def set_dimensions(self, width, height):
        """Sets or updates the width and height of the rectangle with validation."""
        if width <= 0 or height <= 0:
            raise ValueError("Dimensions must be positive values.")
        self.width = width
        self.height = height

    def area(self):
        """Calculates and returns the area of the rectangle."""
        # Check if dimensions have been set before calculation
        if not hasattr(self, 'width') or not hasattr(self, 'height'):
            raise AttributeError("Dimensions must be set using set_dimensions() before calculating area.")
        return self.width * self.height

# --- Verification ---

# 1. Instantiate the object
my_rectangle = Rectangle()

# 2. Set the dimensions dynamically
my_rectangle.set_dimensions(12.5, 4.0)

# 3. Calculate and display the area
print(f"Rectangle Width:  {my_rectangle.width}")
print(f"Rectangle Height: {my_rectangle.height}")
print(f"Calculated Area:  {my_rectangle.area():.2f}")  # Output: 50.00


14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary.
:-
   class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        """Computes base pay purely on hours worked."""
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        # Pass base employee attributes up to the parent constructor
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        """Overrides parent method to append the manager's unique bonus."""
        # Use super() to grab the base salary logic, then add the bonus
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# --- Verification ---

# 1. Create a standard Employee instance
emp = Employee(name="Karan", hours_worked=40, hourly_rate=500)
print(f"Employee: {emp.name} | Total Salary: ₹{emp.calculate_salary():,}")
# Output: Employee: Karan | Total Salary: ₹20,000

# 2. Create a Manager instance
mgr = Manager(name="Priya", hours_worked=40, hourly_rate=800, bonus=15000)
print(f"Manager:  {mgr.name} | Total Salary: ₹{mgr.calculate_salary():,}")
# Output: Manager:  Priya | Total Salary: ₹47,000


15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
calculates the total price of the product.
:-
   class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = float(price)
        self.quantity = int(quantity)

    def total_price(self):
        """Calculates and returns the total valuation of the stock for this product."""
        return self.price * self.quantity

# --- Verification ---

# 1. Instantiate a Product object
item = Product(name="Wireless Mouse", price=1250.00, quantity=4)

# 2. Extract metrics and calculate the total price
print(f"Product Name:    {item.name}")
print(f"Unit Price:      ₹{item.price:,.2f}")
print(f"Stock Quantity:  {item.quantity}")
print(f"Total Value:     ₹{item.total_price():,.2f}")  # Output: ₹5,000.00


16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.
:-
   from abc import ABC, abstractmethod

# 1. Define the Abstract Class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        """Abstract method that must be overridden by all child classes."""
        pass

# 2. Implement the Cow Subclass
class Cow(Animal):
    def sound(self):
        return "Moo!"

# 3. Implement the Sheep Subclass
class Sheep(Animal):
    def sound(self):
        return "Baa!"

# --- Verification ---

# Instantiating the base class directly would cause a TypeError
# generic_animal = Animal() # ❌ This will fail!

# Create child objects
livestock = [Cow(), Sheep()]

# Execute methods polymorphically
for animal in livestock:
    print(f"A {type(animal).__name__} goes: {animal.sound()}")


17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
returns a formatted string with the book's details.
:-
   class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = int(year_published)

    def get_book_info(self):
        """Returns a cleanly formatted reference summary of the book."""
        return f"'{self.title}' by {self.author} ({self.year_published})"

# --- Verification ---

# 1. Instantiate Book objects
book1 = Book(title="The Hobbit", author="J.R.R. Tolkien", year_published=1937)
book2 = Book(title="To Kill a Mockingbird", author="Harper Lee", year_published=1960)

# 2. Retrieve and display the formatted summaries
print(book1.get_book_info())  # Output: 'The Hobbit' by J.R.R. Tolkien (1937)
print(book2.get_book_info())  # Output: 'To Kill a Mockingbird' by Harper Lee (1960)


18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.
:-
   class House:
    def __init__(self, address, price):
        self.address = address
        self.price = float(price)

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        # Forward the core properties up to the House parent constructor
        super().__init__(address, price)
        self.number_of_rooms = int(number_of_rooms)

# --- Verification ---

# 1. Instantiate a standard House object
suburban_home = House(address="123 Maple Street", price=350000)

# 2. Instantiate a Mansion object (which gets all 3 attributes)
luxury_estate = Mansion(address="100 Billionaire Blvd", price=8500000, number_of_rooms=24)

# 3. Print out properties to verify state setup
print("--- Standard House ---")
print(f"Address: {suburban_home.address} | Price: ${suburban_home.price:,.2f}")

print("\n--- Luxury Mansion ---")
print(f"Address: {luxury_estate.address}")
print(f"Price:   ${luxury_estate.price:,.2f}")
print(f"Rooms:   {luxury_estate.number_of_rooms} rooms")

```
